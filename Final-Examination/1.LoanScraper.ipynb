```python
#import Required Libraries
import pandas as pd
from pandas import DataFrame
import numpy as np
import time
from urllib.request import Request, urlopen
from urllib.error import  URLError
```

```python
start_time = time.time()

req = Request('https://www.lendingclub.com/')
try:
    response = urlopen(req)
except URLError as e:
    if hasattr(e, 'reason'):
        print('We failed to reach a server.')
        print('Reason: ', e.reason)
    elif hasattr(e, 'code'):
        print('The server couldn\'t fulfill the request.')
        print('Error code: ', e.code)
else:
        print("Everything is fine")
        
q1 = pd.read_csv('https://lending-club.s3.amazonaws.com/LoanStats_2019Q1.csv')
q2 = pd.read_csv('https://lending-club.s3.amazonaws.com/LoanStats_2019Q2.csv')
q3 = pd.read_csv('https://lending-club.s3.amazonaws.com/LoanStats_2019Q3.csv')
q4 = pd.read_csv('https://lending-club.s3.amazonaws.com/LoanStats_2019Q4.csv')

time.sleep(1)

print('*** Process Completeed ***')
print('Time used:\t',time.time()-start_time)
```

```python
#Merging all sub DataFrame to single DataFrame
loans = pd.concat([q1,q2,q3,q4]) 
loans
```

```python
#Contains info about col and datatype
loans.info(verbose=True, null_counts=True)
```

```python
#Saving it to .csv
loans.to_csv("loan_database_2019.csv")
```
